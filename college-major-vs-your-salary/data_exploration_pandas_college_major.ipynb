# Webscrape Payscale's Website

```python
from bs4 import BeautifulSoup
import requests
```

```python
all_ranks = []
all_majors = []
all_early_pay = []
all_mid_pay = []
all_high_meaning = []

for page_no in range(1, 33):
  payscale_response = requests.get(url=f"https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors/page/{page_no}")
  payscale_response.raise_for_status()
  payscale_soup = BeautifulSoup(payscale_response.text, "html.parser")

  payscale_ranks = payscale_soup.find_all(class_="data-table__cell csr-col--rank")
  payscale_majors = payscale_soup.find_all(class_="data-table__cell csr-col--school-name")
  payscale_metrics =  payscale_soup.find_all(class_="data-table__cell csr-col--right")

  for each_rank in range(len(payscale_ranks)):
    all_ranks.append(int(payscale_ranks[each_rank].find(class_="data-table__value").string))
    all_majors.append(payscale_majors[each_rank].find(class_="data-table__value").string)

  for each_pay in range(0, len(payscale_metrics), 3):
    all_early_pay.append(int(payscale_metrics[each_pay].find(class_="data-table__value").string.replace('$','').replace(',','')))

  for each_pay in range(1, len(payscale_metrics), 3):
    all_mid_pay.append(int(payscale_metrics[each_pay].find(class_="data-table__value").string.replace('$','').replace(',','')))

  for each_pay in range(2, len(payscale_metrics), 3):
    if payscale_metrics[each_pay].find(class_="data-table__value").string == "-":
      all_high_meaning.append(payscale_metrics[each_pay].find(class_="data-table__value").string)
    else:
      all_high_meaning.append(int(payscale_metrics[each_pay].find(class_="data-table__value").string.replace('%','')))
```

```python
payscale_dict = {
    "Rank": all_ranks,
    "Major": all_majors,
    "Early Career Pay": all_early_pay,
    "Mid-Career Pay": all_mid_pay,
    "% High Meaning": all_high_meaning
}
payscale_df = pd.DataFrame(payscale_dict)
payscale_df
```

```python
payscale_df.to_csv("payscale.csv")
```
